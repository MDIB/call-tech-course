```python
def predict(user_weights,optimal_weights):
    weight_sum = 0
    for n in range(len(optimal_weights) - 1):
        weight_sum += user_weights[n] * optimal_weights[n]
    return 1.0 if weight_sum >= 0 else -1.0
```
